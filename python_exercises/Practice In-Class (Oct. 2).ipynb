# Word counter

```python
import requests
```

```python
url = "https://raw.githubusercontent.com/msaxton/18th-century-historians/main/gibbon/gibbon_decline_and_fall/gibbon_decline_volume1_chap01.txt"
```

```python
response = requests.get(url)
```

```python
text = response.text
```

```python
text[:200]
```

```python
word_count = {}
```

```python
# Split() will split strings based on whitespace
words = text.split()
```

```python
words[:100]
```

```python
for word in words:
    lower_word = word.lower()
    if lower_word in word_count.keys():
        word_count[lower_word] += 1
    else:
        word_count[lower_word] = 1
```

```python
for k, v in word_count.items():
    print(k, v)
```

```python
sorted_word_count = sorted(word_count.items(), key=lambda x:x[1], reverse=True)
```

```python
sorted_word_count
```

# Solution 1

```python
# Create list of stop words
remove_words = ["the", "of", "and", "to", "a", "was", "in", "by", "their", "were", "that", "as", "which", 
                "from", "his", "with", "or", "it", "on", "they", "most", "had", "for", "but", "its", "an", 
                "is", "we", "more", "than", "into", "at"]
```

```python
words_no_stops = []
for word in words:   # using original list of words
    word_lower = word.lower()
    if word_lower not in remove_words:
        words_no_stops.append(word_lower)
    
```

```python
word_count2 = {}
```

```python
for word in words_no_stops:
    lower_word = word.lower()
    if lower_word in word_count2.keys():
        word_count2[lower_word] += 1
    else:
        word_count2[lower_word] = 1
```

```python
sorted_word_count2 = sorted(word_count2.items(), key=lambda x:x[1], reverse=True)
```

```python
sorted_word_count2
```
