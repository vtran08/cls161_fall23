# Dictionaries Challenge

## Instructions

Your goal in this challenge is to take a complicated piece of data represented as a Python dictionary and extract specific information from it. You will use geographic data about the city of Troy, the setting of Homer's _Iliad_. The data will be provided for you by running the code below. You can view it outside of the notebook by visiting https://pleiades.stoa.org/places/550595/json. 

Your goal is to use the data to print a sentence that says: "The city of Troy may have been located at 39.9575, 26.238889."
* Hint: the data you are looking for is near the beginning  of the dictionary labeled "coordinates."

## Set up
This code will return a Python dictionary containing the information you need about Troy. The dictionary is called `data`.

```python
# This code will import the data you need and convert it to a dictionary.
# You do not need to understand this code.
import requests  # import code to get information from the web
import json  # import code to change a string to a dictionary
url = 'https://pleiades.stoa.org/places/550595/json'  # url for data
response = requests.get(url)  # this returns a Python object representing the webpage
raw = response.text  # extract the raw text (string) from response object
data = json.loads(raw)  # convert string to dictionary
```

```python
data
```

## Solution

```python
latitude = data['features'][0]['geometry']['coordinates'][1]
```

```python
longitude = data['features'][0]['geometry']['coordinates'][0]
```

```python
print(f"The city of Troy may have been located at {longitude}, {latitude}.")
```
